```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
reconstructed_model = keras.models.load_model("test_model.h5")
```

```python
img = cv2.imread('157362.png')
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
yhat = reconstructed_model.predict(np.expand_dims(resize/255, 0))
```

```python
yhat
```

```python
if yhat > 0.5: 
    print(f'Predicted class is Good')
else:
    print(f'Predicted class is Bad')
```
